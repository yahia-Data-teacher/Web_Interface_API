```python
from flask import Flask
```

```python
app = Flask(__name__)
```

```python
@app.route("/")
def index():
    return "Hello"
```

```python
@app.route("/test1")
def index1():
    return "<h3> Hello world </h3>"
```

```python
@app.route("/test2")
def index2():
    return "<h1> Hello world </h1>"
```

```python
@app.route("/test3")
def index3():
    return """
    
    <h1> Hello world </h1>
    <p>I love python </p>
    
    """
```

```python
app.run(port=5000)
```
